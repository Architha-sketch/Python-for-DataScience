```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/co2_emissions.csv")

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Year"], df["CO2_Emissions_Mt"], marker="o", linestyle="-", color="blue")


plt.xlabel("Year")
plt.ylabel("CO2 Emissions (Million Tonnes)")
plt.title("Global CO2 Emissions Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/student_scores.csv")

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Exam"], df["Score"], marker="o", linestyle="-", color="green")

# Add labels and title
plt.xlabel("Exam Number")
plt.ylabel("Score")
plt.title("Student Performance Over Exams")
plt.grid(True)
plt.tight_layout()
plt.show()
```
